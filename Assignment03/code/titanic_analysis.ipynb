```python
# you would need xlrd - pip install xlrd
t_file = pd.ExcelFile('../data/titanic.xls')
t_df = t_file.parse("titanic", header=None)
t_df.head()
```

### Women and children first?

*** 1. Use the `groupby` method to calculate the proportion of passengers that survived by sex. (25 pts)*** 

```python
#Make the first row of the table into header columns
new_table = t_df.rename(columns=t_df.iloc[0]).loc[1:]
new_table
```

### Calculating Proportions based on the number of survivors:

```python
# We use groupby function to find the number of Males and Females who survived
genderX = new_table[new_table.survived == 1].groupby(by='sex')['sex'].count()

# Then we find the totol number of survivors
genderY = (new_table[new_table.survived == 1])['sex'].count()

# Find the proportion by dividing the number of males and females survivor by the total population
gender = ((genderX/genderY)*100).round(2)
gender.reset_index(name = 'Proportion(%)')
```

*** 2. Calculate the same proportion, but by class and sex. (25 pts)*** 

```python
# We use groupby function to find the number of Males and Females who survived based on their classes
class_sexX = new_table[new_table.survived == 1].groupby(by=['sex','pclass'])['pclass'].count()

# Then we find the totol number of survivors
class_sexY = (new_table[new_table.survived == 1])['pclass'].count()

# Find the proportion by dividing the number of males and females based on their classes survivor by the total population
class_sex = ((class_sexX/class_sexY)*100).round(2)
class_sex.reset_index(name = 'Proportion(%)')
```

*** 3. Create age categories: children (under 14 years), adolescents (14-20), adult (21-64), and senior(65+), and calculate survival proportions by age category, class and sex. (40 pts)***

```python
# Splitting the table of people on the titanic based on age
children = new_table[(new_table['age'] < 14)]
adolescents = new_table[(new_table['age'] >= 14) & (new_table['age'] < 21)]
adult = new_table[(new_table['age'] >= 21) & (new_table['age'] < 65)]
senior = new_table[(new_table['age'] >= 65)]


#Table for survivors who are children
children_catX = children[children.survived == 1].groupby(by=['sex','pclass']).age.count()
children_catY = (new_table[new_table.survived == 1])['age'].count()

children_cat = ((children_catX/children_catY)*100).round(2)


#Table for survivors who are adolescents
adolescents_catX = adolescents[adolescents.survived == 1].groupby(by=['sex','pclass']).age.count()
adolescents_catY = (new_table[new_table.survived == 1])['age'].count()

adolescents_cat = ((adolescents_catX/adolescents_catY)*100).round(2)


#Table for survivors who are adults
adults_catX = adult[adult.survived == 1].groupby(by=['sex','pclass']).age.count()
adults_catY = (new_table[new_table.survived == 1])['age'].count()

adults_cat = ((adults_catX/adults_catY)*100).round(2)


#Table for survivors who are seniors
seniors_catX = senior[senior.survived == 1].groupby(by=['sex','pclass']).age.count()
seniors_catY = (new_table[new_table.survived == 1])['age'].count()

seniors_cat = ((seniors_catX/seniors_catY)*100).round(2)

# total_survivors
total_survivor = (new_table[new_table.survived == 1])['sex'].count()

# Printing out the number of surviors of each age category
print("children: ", sum(children_catX))
print("adolescents: ", sum(adolescents_catX))
print("adult: ", sum(adults_catX))
print("senior: ", sum(seniors_catX))
print("Total surviors: ", total_survivor)

```

### Displaying the Survival Proportions based on sex, class, and age categories:

```python
# First we merge all the various dataframes together so all information can be seen on one big dataframe
merge1 = pd.merge(children_cat,adolescents_cat,on=['sex','pclass'])
merge1.rename(columns = {'age_x':'Children_Proportion(%)','age_y':'Adolescents_Proportion(%)'}, inplace = True)
merge1

merge2 = pd.merge(merge1, adults_cat, on=['sex','pclass'])
merge2.rename(columns = {'age':'Adults_Proportion(%)'}, inplace = True)
merge2

final_data = pd.merge(merge2, seniors_cat, on=['sex','pclass'], how='left')
final_data = final_data.fillna(0)
final_data.rename(columns = {'age':'Seniors_Proportion(%)'}, inplace = True)


final_data
```
